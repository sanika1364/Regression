# Regression And Its Evaluation



**Question 1: What is Simple Linear Regression?**

Simple Linear Regression is a statistical method used to model the relationship between two continuous variables: a dependent variable (the one you want to predict) and an independent variable (the one used for prediction). It assumes a linear relationship, meaning the relationship can be represented by a straight line. The goal is to find the line that best fits the data, minimizing the difference between the observed values of the dependent variable and the values predicted by the line.

**Question 2: What are the key assumptions of Simple Linear Regression?**

Here are the key assumptions of Simple Linear Regression:

*   **Linearity:** The relationship between the independent and dependent variables is linear.
*   **Independence:** The observations are independent of each other.
*   **Homoscedasticity:** The variance of the residuals is constant across all levels of the independent variable.
*   **Normality:** The residuals are normally distributed.

**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**

**Heteroscedasticity** is the violation of the homoscedasticity assumption in regression analysis. It occurs when the variance of the residuals (the differences between observed and predicted values) is not constant across all levels of the independent variable. In other words, the spread of the residuals changes as the value of the independent variable changes.

**Why it is important to address:**

Heteroscedasticity can have several negative consequences for your regression model:

*   **Biased standard errors:** This leads to incorrect p-values and confidence intervals, making it difficult to determine the statistical significance of your predictors.
*   **Inefficient estimates:** The regression coefficients may still be unbiased, but they will not be the most efficient estimates possible, meaning there are other possible sets of coefficients that would produce a lower variance.
*   **Unreliable hypothesis testing:** Because the standard errors are biased, hypothesis tests about the coefficients will be unreliable.
*   **Misleading model evaluation:** Goodness-of-fit measures like R-squared can be misleading in the presence of heteroscedasticity.

Addressing heteroscedasticity helps ensure that your regression model provides accurate and reliable results.

**Question 4: What is Multiple Linear Regression?**

Multiple Linear Regression is a statistical method used to model the relationship between a dependent variable and two or more independent variables. It's an extension of Simple Linear Regression, allowing you to consider the influence of multiple factors simultaneously when making predictions. The goal is to find the linear equation that best fits the data by minimizing the sum of the squared differences between the observed and predicted values.

**Question 5: What is polynomial regression, and how does it differ from linear regression?**

**Polynomial Regression** is a form of regression analysis in which the relationship between the independent variable ($x$) and the dependent variable ($y$) is modeled as an *n*-th degree polynomial. While linear regression models the relationship with a straight line, polynomial regression models it with a curved line.

Here's how it differs from linear regression:

*   **Relationship modeled:** Linear regression models a linear relationship ($y = \beta_0 + \beta_1x$), while polynomial regression models a non-linear relationship by including polynomial terms of the independent variable ($y = \beta_0 + \beta_1x + \beta_2x^2 + ... + \beta_nx^n$).
*   **Flexibility:** Polynomial regression is more flexible than linear regression and can fit a wider range of data patterns, including curves and bends.
*   **Model complexity:** As the degree of the polynomial increases, the model becomes more complex. This increased complexity can lead to overfitting if not managed carefully.
*   **Interpretation:** The interpretation of coefficients in polynomial regression is less straightforward than in linear regression, as the coefficients represent the change in the dependent variable for a unit change in the corresponding polynomial term, not just the independent variable itself.

In essence, polynomial regression allows you to capture more complex relationships between variables than simple or multiple linear regression, but it comes with the trade-off of increased model complexity and potential for overfitting.

**Calculate vif**

**Subtask:**
Calculate the Variance Inflation Factor (VIF) for each independent variable to check for multicollinearity.

**Reasoning**:
Calculate the Variance Inflation Factor (VIF) for each independent variable to check for multicollinearity.


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Report the results
display(vif_data)
```

**Report vif results**

**Subtask:**
Report the calculated VIF values to indicate the presence and extent of multicollinearity.

**Reasoning**:
Display the calculated VIF values and explain their meaning in the context of multicollinearity.


```python
# Display the VIF data
display(vif_data)

# Explain the VIF values
print("VIF values indicate the extent of multicollinearity. A VIF value above 5 or 10 is generally considered to indicate significant multicollinearity.")
print("\nComment on the VIF values:")
print(f"The VIF value for 'Area' is {vif_data[vif_data['feature'] == 'Area']['VIF'].iloc[0]:.2f}.")
print(f"The VIF value for 'Rooms' is {vif_data[vif_data['feature'] == 'Rooms']['VIF'].iloc[0]:.2f}.")
print("Both 'Area' and 'Rooms' have very high VIF values (above 10), suggesting a strong linear relationship between these two features and indicating significant multicollinearity.")
```

**Summary:**

**Data Analysis Key Findings**

*   Both independent variables have very high VIF values (above the generally accepted threshold of 5 or 10), indicating significant multicollinearity between 'Area' and 'Rooms'.

**Insights or Next Steps**

**Question 8: Implement polynomial regression on the following data:**

● X = [1, 2, 3, 4, 5]

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

**Fit a 2nd-degree polynomial and plot the resulting curve.**

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Prepare the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])
```

```python
# Create polynomial features (2nd degree)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
```

```python
# Create and train the linear regression model on the polynomial features
model = LinearRegression()
model.fit(X_poly, Y)
```

```python
# Generate points for plotting the polynomial curve
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot = model.predict(X_plot_poly)
```

```python
# Plot the original data points and the polynomial regression curve
plt.scatter(X, Y, color='blue', label='Original data')
plt.plot(X_plot, Y_plot, color='red', label='2nd-degree Polynomial Regression')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

**Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.**

To address heteroscedasticity and multicollinearity in a regression model for predicting house prices, you would take the following steps:

**Addressing Heteroscedasticity:**

1.  **Identify Heteroscedasticity:**
    *   **Residuals Plot:** As demonstrated in the previous question, create a scatter plot of residuals versus the predicted values or the independent variables. If the spread of residuals changes systematically across the plot (e.g., widens or narrows), heteroscedasticity is present.
    *   **Statistical Tests:** Conduct formal statistical tests like the Breusch-Pagan test or the White test to confirm the presence and significance of heteroscedasticity.

2.  **Address Heteroscedasticity:**
    *   **Data Transformation:** Apply a transformation to the dependent variable (house price) or the independent variables. Common transformations include taking the logarithm, square root, or reciprocal of the variable. This can help stabilize the variance of the residuals.
    *   **Weighted Least Squares (WLS):** If the form of heteroscedasticity is known or can be estimated, you can use Weighted Least Squares regression. WLS assigns different weights to observations based on their variance, giving less weight to observations with higher variance.
    *   **Robust Standard Errors:** Use robust standard errors (also known as White standard errors or heteroscedasticity-consistent standard errors). These standard errors are adjusted to account for heteroscedasticity, providing more reliable p-values and confidence intervals even when the assumption of homoscedasticity is violated.
    *   **Model Respecification:** Consider if there are any missing important variables or if the functional form of the model is incorrect. Adding relevant variables or using a different model (e.g., a non-linear model) might alleviate heteroscedasticity.

**Addressing Multicollinearity:**

1.  **Identify Multicollinearity:**
    *   **Correlation Matrix:** Calculate the correlation matrix for your independent variables. High correlation coefficients (e.g., above 0.7 or 0.8 in absolute value) between independent variables suggest potential multicollinearity.
    *   **Variance Inflation Factor (VIF):** As demonstrated in Question 7, calculate the VIF for each independent variable. VIF measures how much the variance of the estimated regression coefficient is increased due to multicollinearity. VIF values above 5 or 10 are generally considered indicative of significant multicollinearity.

2.  **Address Multicollinearity:**
    *   **Remove Highly Correlated Variables:** If two or more independent variables are highly correlated and theoretically represent similar information, consider removing one of them from the model. Choose the variable that is less theoretically important or has a weaker relationship with the dependent variable.
    *   **Combine Correlated Variables:** Create a new composite variable by combining the highly correlated variables. For example, if 'Area' and 'Number of Rooms' are highly correlated, you could create a new variable like 'Area per Room'.
    *   **Dimensionality Reduction Techniques:** Techniques like Principal Component Analysis (PCA) can be used to reduce the number of independent variables while retaining most of the variance in the data. The principal components are uncorrelated and can be used as predictors in the regression model.
    *   **Ridge Regression or Lasso Regression:** These are regularized regression techniques that can handle multicollinearity. Ridge regression adds a penalty term to the sum of squared errors that shrinks the regression coefficients, while Lasso regression can shrink some coefficients to zero, effectively performing variable selection.

**Ensuring a Robust Model:**

After addressing heteroscedasticity and multicollinearity, you would:

*   **Re-evaluate the Model:** Fit the regression model again with the chosen strategies for addressing heteroscedasticity and multicollinearity.
*   **Check Assumptions:** Re-examine the assumptions of the regression model (linearity, independence, normality of residuals) to ensure they are met.
*   **Evaluate Model Performance:** Assess the performance of the revised model using appropriate evaluation metrics (e.g., R-squared, adjusted R-squared, Mean Squared Error, Root Mean Squared Error).
*   **Interpret Results:** Interpret the coefficients of the revised model, keeping in mind the impact of the steps taken to address the issues.

By systematically identifying and addressing heteroscedasticity and multicollinearity, you can build a more robust and reliable model for predicting house prices.
